### Zadanie Passengers
Zbiór danych "passengers" z biblioteki seaborn zawiera informacje dotyczące liczby pasażerów podróżujących samolotem w różnych miesiącach i latach.

Struktura zbioru danych:

* Kolumna "year" zawiera informacje o roku, do którego odnoszą się dane.
* Kolumna "month" zawiera informacje o miesiącu, do którego odnoszą się dane.
* Kolumna "passengers" zawiera informacje o liczbie pasażerów podróżujących samolotem w danym miesiącu i roku.

Dane "passengers" zawierają informacje o liczbie pasażerów, którzy podróżowali samolotem w poszczególnych miesiącach i latach. Twoim zadaniem jest przygotowanie raportu używając Pandas, który odpowie na następujące pytania:

* Jak wygląda ogólny trend wzrostu liczby pasażerów w czasie, użyj .plot()?

* W którym roku i miesiącu zanotowano największą liczbę pasażerów?

* Które miesiące i lata mają tendencję do wyższej niż średnia liczby pasażerów?

* Czy istnieją okresy, w których liczba pasażerów jest wyraźnie niższa niż przeciętna?

* Jakie są różnice w liczbie pasażerów między różnymi latami? wskazówka: metoda .diff()

```python
import pandas as pd 

passengers = pd.read_csv('/kaggle/input/seaborn-practice/flights.csv')
passengers.head()
```

1. Jak wygląda ogólny trend wzrostu liczby pasażerów w czasie?

```python
passengers.plot(y='passengers')
```

2. W którym roku i miesiącu zanotowano największą liczbę pasażerów?


```python
passengers[passengers['passengers']==passengers['passengers'].max()]
```

3. Które miesiące i lata mają tendencję do wyższej niż średnia liczby pasażerów?


```python
passengers[passengers['passengers']>passengers['passengers'].mean()]
```

4. Czy istnieją okresy, w których liczba pasażerów jest wyraźnie niższa niż przeciętna?


```python
#wyraznie nizsza niz przecieta -> zawierajace sie w kwantylu <0.5 - tu 0.3, zeby nie wyszlo za duzo wynikow
passengers[passengers['passengers']<passengers['passengers'].quantile(0.3)]
```

5. Jakie są różnice w liczbie pasażerów między różnymi latami? wskazówka: metoda .diff()

```python
passengers['passengers'].diff()
```

### Zadanie penguins

Zbiór danych "penguins" z biblioteki seaborn zawiera informacje dotyczące pingwinów. Ten zbiór danych zawiera pomiary różnych cech fizycznych pingwinów, takich jak gatunek, płeć, długość skrzydła, długość dzioba, masa ciała itp.

Struktura zbioru danych "penguins":

* Kolumna "species" zawiera informacje o gatunku pingwina, przyjmując jedną z trzech wartości: "Adelie", "Chinstrap" lub "Gentoo".
* Kolumna "island" zawiera informacje o wyspie, na której mieszkają pingwiny, przyjmując jedną z trzech wartości: "Biscoe", "Dream" lub "Torgersen".
* Kolumna "bill_length_mm" zawiera pomiary długości dzioba w milimetrach.
* Kolumna "bill_depth_mm" zawiera pomiary głębokości dzioba w milimetrach.
* Kolumna "flipper_length_mm" zawiera pomiary długości skrzydła (płetwy) w milimetrach.
* Kolumna "body_mass_g" zawiera pomiary masy ciała pingwina w gramach.
* Kolumna "sex" zawiera informacje o płci pingwina, przyjmując jedną z dwóch wartości: "male" lub "female".

```python
import pandas as pd 

penguins = pd.read_csv('/kaggle/input/seaborn-practice/penguins.csv').dropna()
penguins.head()
```

Analiza danych dotyczących pingwinów

Zbiór danych "penguins" z biblioteki seaborn zawiera informacje o różnych cechach pingwinów, takich jak gatunek, płeć, długość skrzydła, długość dzioba itp. Twoim zadaniem jest przygotowanie raportu, który odpowie na następujące pytania:



1. Która z trzech gatunków pingwinów jest najliczniejsza w zbiorze danych?

2. Jakie są średnie wartości poszczególnych cech dla każdego gatunku pingwina?

3. Oblicz średnią masę ciała pingwinów dla każdego gatunku osobno.

4. Porównaj rozkład długości skrzydeł pingwinów na poszczególnych wyspach.

5. Znajdź najmniejszą i największą wartość długości dzioba dla każdej płci.

6. Oblicz sumę mas ciała pingwinów dla każdej kombinacji gatunku i płci.

7. Zgrupuj pingwiny na podstawie gatunku i wyspy, a następnie oblicz medianę masy ciała dla każdej grupy.

8. Oblicz średnią długość skrzydeł pingwinów dla każdego roku, uwzględniając tylko gatunek Adelie.

9. Znajdź najliczniejszą kombinację gatunku, wyspy i płci w zbiorze danych "penguins".

1. Która z trzech gatunków pingwinów jest najliczniejsza w zbiorze danych?

```python
#penguins['species'].value_counts().idxmax()
penguins.groupby('species').size()
```

2. Jakie są średnie wartości poszczególnych cech dla każdego gatunku pingwina?


```python
#Wyspa i plec nie jest dana z ktorej mozna wyznaczyc srednia
penguins.groupby('species').agg({'bill_length_mm':'mean','bill_depth_mm':'mean','flipper_length_mm':'mean','body_mass_g':'mean'})
```

3. Oblicz średnią masę ciała pingwinów dla każdego gatunku osobno.


```python
penguins.groupby('species')['body_mass_g'].mean()
```

4. Porównaj rozkład długości skrzydeł pingwinów na poszczególnych wyspach.

```python
penguins.groupby('species')['flipper_length_mm'].mean()
```

5. Znajdź najmniejszą i największą wartość długości dzioba dla każdej płci.


```python
penguins.groupby('sex')['bill_length_mm'].agg(['min', 'max'])
```

6. Oblicz sumę mas ciała pingwinów dla każdej kombinacji gatunku i płci.


```python
penguins.groupby(['species','sex'])['body_mass_g'].sum()
```

7. Zgrupuj pingwiny na podstawie gatunku i wyspy, a następnie oblicz medianę masy ciała dla każdej grupy.


```python
penguins.groupby(['species','island'])['body_mass_g'].quantile(0.5)
```

8. Oblicz średnią długość skrzydeł pingwinów dla każdej płci, uwzględniając tylko gatunek Adelie.


```python
penguins[penguins['species']=='Adelie'].groupby('sex')['flipper_length_mm'].mean()
```

9. Znajdź najliczniejszą kombinację gatunku, wyspy i płci w zbiorze danych "penguins".

```python
penguins.groupby(['species','island','sex']).size().idxmax()
```
